```python
def do_q1(x, y, axn):
    # change x to a row vector (nparray) and y to a column vec (nparray)
    x_fixed = np.array(x).reshape(-1, 1)
    y_fixed = np.array(y)

    # do the regression
    model = LinearRegression().fit(x_fixed, y_fixed)

    # create the fit line
    m = model.coef_
    b = model.intercept_
    fit = m * x + b

    # calculate r^2
    r2, p = pearsonr(x, y)

    # plot results
    ax[axn].scatter(x, y, color="darkred", alpha=0.6)
    ax[axn].plot(x, fit, color="k")
    ax[axn].annotate(f"$R^2 =${round(r2,4)}", [0.6 * max(x), 0.8 * min(y)])
    ax[axn].annotate(
        f"$y = {np.round(m,3)[0]}x + {np.round(b,3)}$", [0.6 * max(x), min(y)]
    )


fig, ax = plt.subplots(3, figsize=(10, 10))
do_q1(corr["x"], corr["y"], 0)
do_q1(corr["x2"], corr["y2"], 1)
do_q1(corr["x3"], corr["y3"], 2)

ax[0].set_title("Data Set 1")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

ax[1].set_title("Data Set 2")
ax[1].set_xlabel("x2")
ax[1].set_ylabel("y2")

ax[2].set_title("Data Set 3")
ax[2].set_xlabel("x3")
ax[2].set_ylabel("y3")

plt.tight_layout()
```

```python
### QUESTION 2 ###

# get data and do Q&D plot
mlr = pd.read_csv("MLR.csv")
X = mlr[["x1","x2","x3","x4","x5","x6"]]
Y = mlr[["y"]]
mlr.plot(subplots=True, figsize=(15,8), xlabel="time (arbitrary units)");
```

```python
# do MLR just like tutorial 2 - copy Sam/Cole's code
model = LinearRegression().fit(X, Y)
ypred_MLR = model.predict(X)  # y predicted by MLR
intercept_MLR = model.intercept_[0]  # intercept predicted by MLR
coef_MLR = model.coef_.flatten()  # regression coefficients in MLR model
R2_MLR = model.score(X, Y)  # R-squared value from MLR model

Y_mod = (
    intercept_MLR
    + (coef_MLR[0] * X["x1"])
    + (coef_MLR[1] * X["x2"])
    + (coef_MLR[2] * X["x3"])
    + (coef_MLR[3] * X["x4"])
    + (coef_MLR[4] * X["x5"])
    + (coef_MLR[5] * X["x6"])
)

print("MLR results:")
print("a0 = " + str(intercept_MLR))
print("a1 = " + str(coef_MLR[0]))
print("a2 = " + str(coef_MLR[1]))
print("a3 = " + str(coef_MLR[2]))
print("a4 = " + str(coef_MLR[3]))
print("a5 = " + str(coef_MLR[4]))
print("a6 = " + str(coef_MLR[5]))
print("R2 = " + str(R2_MLR))
```

```python
fig, ax = plt.subplots(1,2, figsize = (15,5))

Y.plot(ax=ax[0], label="measured Y")
Y_mod.plot(ax=ax[0], label="modeled Y")
ax[0].set_xlabel("time (arbitrary units)")

ax[0].legend()

ax[1].scatter(Y,Y_mod, alpha=0.6)
ax[1].annotate(f"$R^2$: {R2_MLR}", (10,5))
ax[1].set_xlabel("measured Y")
ax[1].set_ylabel("modeled Y")
```
